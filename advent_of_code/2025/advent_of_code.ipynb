### Day 1
#### Secret Entrance 
##### part 1
A safe has a dial with only an arrow on it; around the dial are the numbers 0 through 99 in order. As you turn the dial, it makes a small click noise as it reaches each number.
The puzzle input contains a sequence of rotations
A rotation starts with an:
L = toward lower numbers
R = toward higher numbers
Then, the rotation has a distance value

The dial starts by pointing at 50

The actual password is the number of times the dial is left pointing at 0 after any rotation in the sequence.

##### part 2
you're actually supposed to count the number of times any click causes the dial to point at 0, regardless of whether it happens during a rotation or at the end of one.


```python
# Get the data
test = False

if test:
    file = 'test_day1.txt'
else:
    file = 'day1.txt'
rotations = []  
with open(file) as data:
    for line in data:
        rotations.append(line.strip())

# Process the data part 1.1
land_on_zero_counter = 0
zero_counter = 0
current_number = 50

for rotation in rotations:
    direction = rotation[0]
    steps = int(rotation[1:])
    for i in range(steps):
        if direction == 'L':
            current_number -= 1
            if current_number < 0:
                current_number = 99
        else:
            current_number += 1
            if current_number > 99:
                current_number = 0
        if current_number == 0:
            zero_counter += 1
        
    if current_number == 0:
        land_on_zero_counter += 1

print(f"Day 1.1 land on zero counter:{land_on_zero_counter}")
print(f"Day 1.2 number of zeros:{zero_counter}")
```

### Day 2
#### Gift Shop 
##### part 1
One of the younger Elves was playing on a gift shop computer and managed to add a whole bunch of invalid product IDs.
A few product ID ranges (your puzzle input) that you'll need to check.
The ranges are separated by commas (,); each range gives its first ID and last ID separated by a dash (-).
invalid IDs is any ID which is made only of some sequence of digits repeated twice.
(55, 6464, 123123 etc) or any ID that starts with 0

What do you get if you add up all of the invalid IDs?

##### part 2
Now, an ID is invalid if it is made only of some sequence of digits repeated **at least** twice.

```python
# Get the data
test = False

if test:
    file = 'test_day2.txt'
else:
    file = 'day2.txt'

# Counter for results
first_invalidIDs_added_up = 0
second_invalidIDs_added_up = 0

# Open data file
with open(file) as f:
    # split between instances
    data = f.read().split(',')
    for line in data:
        # Split ranges in start and end
        range_start, range_end = line.split('-')
        range_start, range_end = int(range_start), int(range_end)

        # Loop through all productIDs within current range
        for prod_ID in range(range_start, range_end + 1):
            # 2.1
            # Split productID in 2 halves
            first_halve, second_halve = str(prod_ID)[:len(str(prod_ID))//2], str(prod_ID)[len(str(prod_ID))//2:]
            # If both halves are similar, add productID to counter
            if first_halve == second_halve:
                first_invalidIDs_added_up += prod_ID

            # 2.2
            for i in range(len(str(prod_ID))// 2, 0, -1):
                digits = [str(prod_ID)[k:k+i] for k in range(0,len(str(prod_ID)),i)]
                if all(x == digits[0] for x in digits):
                    second_invalidIDs_added_up += prod_ID
                    break

print(f"Day 2.1 sum of invalid IDs:{first_invalidIDs_added_up}")
print(f"Day 2.2 sum of invalid IDs:{second_invalidIDs_added_up}")
```

### Day 3
#### Lobby
##### part 1
Batteries each labeled with their joltage rating, a value from 1 to 9 = your puzzle input
Each line of digits in your input corresponds to a single bank of batteries.

Within each bank, you need to turn on exactly two batteries; the joltage that the bank produces is equal to the number formed by the digits on the batteries you've turned on. For example, if you have a bank like 12345 and you turn on batteries 2 and 4, the bank would produce 24 jolts. (You cannot rearrange batteries.)

find the largest possible joltage
total output joltage is the sum of the maximum joltage from each bank

##### part 2
Now, you need to make the largest joltage by turning on exactly twelve batteries within each bank.


```python
# Get the data
test = False
number_of_batteries = 12

if test:
    file = 'test_day3.txt'
else:
    file = 'day3.txt'

# sum joltage
sum_joltage = 0

# Open data file
with open(file) as data:
    for line in data:
        line = line.strip()
        joltage = ''
        previous_j = -1
        # Loop 12x through line
        for i in range(number_of_batteries):
            # Find the largest digit 
            largest_digit = 0
            for j, digit in enumerate(line):
                # Between previous largest and how many numbers still have to be found after this
                if previous_j < j < len(line)-(number_of_batteries - 1-i):
                    if int(digit) > largest_digit:
                        largest_digit = int(digit)
                        # Set new previous largest location for next loop
                        previous_j = j
            # Add joltage to current bank
            joltage = joltage + str(largest_digit)
        # Add this current bank to total joltage
        sum_joltage += int(joltage)
print(sum_joltage)
```

### Day 4
#### Printing Department
##### part 1
Rolls of paper (@) are arranged on a large grid, puzzle input.
Forklifts can only access a roll of paper if there are fewer than four rolls of paper in the eight adjacent positions.
How many rolls of paper can be accessed by a forklift?
##### part 2
Once a roll of paper can be accessed by a forklift, it can be removed.
How many total rolls of paper could the Elves remove if they keep repeating this process?

```python
# Get the data
test = False

if test:
    file = 'test_day4.txt'
else:
    file = 'day4.txt'

# Number of accessible rolls
accessible_rolls = 0

# Open data file and reconstruct the grid
grid = []
with open(file) as data:
    for line in data:
        row = []
        for position in line.strip():
            row.append(position)
        grid.append(row)

def check_neighbours(grid, current_row, current_position):
    neighbouring_rolls = 0
    # Define neighbours
    neighbours_dict = {"upperleft": [-1, -1],
                    "uppermid": [0,-1],
                    "upperright": [1,-1],
                    "left": [-1, 0],
                    "right": [1, 0],
                    "lowerleft": [-1, 1],
                    "lowermid": [0, 1],
                    "lowerright": [1, 1]}
    # Loop through all neighbours
    for neighbour in neighbours_dict:
        grid_move, row_move= neighbours_dict[neighbour]
        neighbour_row = current_row + grid_move
        neighbour_position = current_position + row_move
        # If neighbour is inside grid
        if 0 <= neighbour_row < len(grid) and 0 <= neighbour_position < len(grid[current_row]):
            # Check if neighbour is a roll
            neighbour_ID = grid[neighbour_row][neighbour_position]
            if neighbour_ID == '@':
                neighbouring_rolls += 1   
    return neighbouring_rolls
        
# Loop through grid
removed_rolls = 0
remove = 1
while remove > 0:
    to_be_removed = []
    for current_row, row in enumerate(grid):
        for current_position, position in enumerate(row):           
                # Check all rolls
                if position == "@":
                    # If roll, then check if less than 4 neighbours are rolls
                    if check_neighbours(grid, current_row, current_position) < 4:
                        accessible_rolls += 1
                        # Add to be removed list
                        to_be_removed.append([current_row,current_position])
    # Remove accessible_rolls from grid
    removed_rolls += len(to_be_removed)

    for roll in to_be_removed:
        grid[roll[0]][roll[1]] = '.'

    remove = len(to_be_removed)
print(removed_rolls)
```

### Day 5
#### Cafeteria
##### part 1
Because of their complicated new inventory management system, they can't figure out which of their ingredients are fresh and which are spoiled. They give you a copy of their database (your puzzle input).
It consists of a list of fresh ingredient ID ranges, a blank line, and a list of available ingredient IDs.
How many of the available ingredient IDs are fresh?
##### part 2
How many ingredient IDs are considered to be fresh according to the fresh ingredient ID ranges?

```python
# Get the data
test = False

if test:
    file = 'test_day5.txt'
else:
    file = 'day5.txt'

# Make lists to append fresh and available ingredients to
available_fresh_ingredient_counter = 0
fresh_ingredients = []
available_ingredients = []
# Loop through database
with open(file) as data:
    for line in data:
        # Extract fresh ingredients ranges
        if '-' in line:
            x, y = line.strip().split('-')
            fresh_ingredients.append([int(x), int(y)])
        # Append available ingredients to list
        else:            
            if line.strip().isdigit():
                available_ingredients.append(int(line.strip()))

# Loop through al available ingredients
for ingredientID in available_ingredients:
    # Check if in any of the ranges
    for IDrange in fresh_ingredients:
        if int(IDrange[0]) <= ingredientID <= int(IDrange[1]):
            available_fresh_ingredient_counter +=1
            break
print(str(available_fresh_ingredient_counter))

# Check everything in ranges
# Sort ranges
fresh_ingredients.sort()
# Use first range as starting point
current_low, current_high = fresh_ingredients[0]
total_num_of_fresh_ids = 0
# Loop through all ranges
for next_low, next_high in fresh_ingredients:
    # If new range_start is lower or equal to previous highest range_end
    if next_low <= current_high:
        # Update current range end
        current_high = max(current_high, next_high)
    # If new range_start is higher than previous highest range_end
    else:
        # add all IDs in new range to total
        total_num_of_fresh_ids += current_high - current_low + 1
        # update current range start and end
        current_low, current_high = next_low, next_high

# Add the final missing IDs
total_num_of_fresh_ids += current_high - current_low + 1
print(total_num_of_fresh_ids)
```

### Day 6
#### Trash Compactor
##### part 1
**Cephalopod Math homework**
The math worksheet (your puzzle input) consists of a list of problems; 
each problem has a group of numbers that need to be either added (+) or multiplied (*) together.
Each problem's numbers are arranged vertically; at the bottom of the problem is the symbol for the operation
What is the grand total found by adding together all of the answers to the individual problems?
##### part 2
Cephalopod math is written right-to-left in columns. 
Each number is given in its own column, with the most significant digit at the top and the least significant digit at the bottom.


```python
# Get the data
test = False

if test:
    file = 'test_day6.txt'
else:
    file = 'day6.txt'

# Import pandas to be able to make a df
import pandas as pd
data = pd.read_csv(file, delimiter=r"\s+", header=None) # delimiter=r"\s+" for varying number of spaces

# Loop through problems
answer_all_problems_combined = 0
for column in data.columns:
    # Make into a list
    problem = data[column].tolist()
    # Get symbol for operation
    operator = problem[-1]
    # Start with first number
    answer_current_problem = int(problem[0])

    # loop through numbers, except for first and the operator
    for number in problem[1:-1]:
        # Do the calculation
        answer_current_problem = eval(str(answer_current_problem) + operator + number)
    
    # Add answer to all other answers
    answer_all_problems_combined += answer_current_problem
print(answer_all_problems_combined)

# part 2, spaces matter, so no df
# Get lines, without enters
lines = [list(line.rstrip('\n')) for line in open(file)]

answer_all_problems_combined = 0
# Set start operator
operator = lines[-1][-1]
problem = []

# Loop through vertical lines
for i in range(1, len(lines[0]) + 1):
    # Start with clear number varable
    number = ''
    # Loop through horizontal lines and get correct numbers
    for line in lines:
        character = line[-i]
        # Get rid of spaces now
        if character.isdigit():
            number += character
        # Update operator when necesarry
        if character == '+' or character == '*':
            operator = character
    # After each problem, solve the problem
    if len(number) == 0:
        # Start with first number
        answer_current_problem = int(problem[0])
        # Loop through numbers, except for first
        for digit in problem[1:]:
            answer_current_problem = eval(str(answer_current_problem) + operator + digit)
        answer_all_problems_combined += answer_current_problem
        # Reset the problem
        problem = []
    # If not end of problem, append number to problem
    else:
        problem.append(number)
# Add final problem
# Start with first number
answer_current_problem = int(problem[0])
# Loop through numbers, except for first
for digit in problem[1:]:
    answer_current_problem = eval(str(answer_current_problem) + operator + digit)
answer_all_problems_combined += answer_current_problem

print(answer_all_problems_combined)
```

### Day 7
#### Laboratories
##### part 1
There's an issue with one of the tachyon manifolds. You have a diagram of the tachyon manifold (your puzzle input).
A tachyon beam enters the manifold at the location marked S; tachyon beams always move downward. Tachyon beams pass freely through empty space (.). However, if a tachyon beam encounters a splitter (^), the beam is stopped; instead, a new tachyon beam continues from the immediate left and from the immediate right of the splitter.

How many times will the beam be split?
##### part 2
It isn't a classical tachyon manifold - it's a quantum tachyon manifold.
Each time a particle reaches a splitter, it's actually time itself which splits.
In one timeline, the particle went left, and in the other timeline, the particle went right.
In total, how many different timelines would a single tachyon particle end up on?

```python
# Get the data
test = False

if test:
    file = 'test_day7.txt'
else:
    file = 'day7.txt'
lines = [list(line.rstrip('\n')) for line in open(file)]

# Set a counter for beam splits
splits = 0

# Grid dimensions:
grid_length = len(lines)
# Loop through lines:
for i, line in enumerate(lines):
    if i < grid_length - 1:
        # Search for Startpoint or beams in current line
        for j, char in enumerate(line):
            if char == 'S':
                char, lines[i][j] = '1','1'
            # If there is a beam (number)
            if char.isdigit():
                # Check line below
                down_char = lines[i +1][j]
                # If there is empty space, let the beam continue
                if down_char == '.':
                    lines[i +1][j] = char
                # If there is a beam down, add it up
                if down_char.isdigit():
                    lines[i +1][j] = str(int(char) + int(down_char))
                # If there is a splitter, split the beam down, if not outside of grid
                if down_char == '^':
                    # Check next to beam
                    left = lines[i+1][j-1]
                    right = lines[i+1][j+1]
                    # if empty space, let the beam continue
                    if left == '.':
                        lines[i+1][j-1] = char
                    if right == '.':
                        lines[i+1][j+1] = char
                    # if there is a beam already, add it up
                    if left.isdigit():
                        lines[i+1][j-1] = str(int(char) + int(left))
                    if right.isdigit():
                        lines[i+1][j+1] = str(int(char) + int(right))
                    # Add split to splits counter
                    splits += 1
# Count different timelines
timelines = 0
for char in lines[-1]:
    if char.isdigit():
        timelines += int(char)

print('part 1: ' + str(splits))
print('part 2: ' + str(timelines))
```

### Day 8
#### Playground
##### part 1
Connect small electrical junction boxes with long strings of lights.
Most of the junction boxes don't provide electricity; however, when two junction boxes are connected by a string of lights, electricity can pass between those two junction boxes.
Figure out which junction boxes to connect so that electricity can reach every junction box. 
A list of all of the junction boxes' positions in 3D space (your puzzle input).
Position is given as X,Y,Z coordinates, one junction box per line.
Connect pairs of junction boxes that are as close together as possible according to straight-line distance (Euclidean distance or Pythagorean distance).
distance(p,q)= sqrt {(p1-q1)**2+(p2-q2)**2+(p3-q3)**2}
distance ^2 = (p1-q1)**2 + (p2-q2)**2 + (p3-q3)**2
connect together the 1000 pairs of junction boxes which are closest together. 
Afterward, what do you get if you multiply together the sizes of the three largest circuits?

```python
# Get the data
test = True
if test == True:
    pairing = 10
else:
    pairing = 1000

if test:
    file = 'test_day8.txt'
else:
    file = 'day8.txt'
lines = [list(line.rstrip('\n').split(',')) for line in open(file)]

import math

shortest_distances = []
distances = {}

for i, LINE in enumerate(lines):
    # Get coordinates of first to compare
    X,Y,Z = LINE
    # Find closest junction box
    for j in range(i+1, len(lines)): # only 1 way
        # Get coordinates of second to compare
        x,y,z = lines[j]
        # Calculate distance
        distance = math.sqrt((int(X)-int(x))**2 + (int(Y)-int(y))**2 + (int(Z)-int(z))**2)
        # Add distance to dict
        distances[i,j] = distance

# Find shortest distances
for i in range(pairing):
    shortest_distance = min(distances, key=distances.get)
    shortest_distances.append(shortest_distance)
    del distances[shortest_distance]


# Make circuits
circuits = []
for i,j in shortest_distances:
    # Find all circuits that contain i or j
    matching_circuits = []
    for c, circuit in enumerate(circuits):
        if i in circuit or j in circuit:
            matching_circuits.append(c)
    # Neither point is in any circuit, create new one
    if len(matching_circuits) == 0:
        circuits.append([i,j])
    # One circuit found, add the missing point
    elif len(matching_circuits) == 1:
        c = matching_circuits[0]
        if i not in circuits[c]:
            circuits[c].append(i)
        if j not in circuits[c]:
            circuits[c].append(j)
    # Multiple circuits found, merge them all
    # Add both points to the first matching circuit
    else:
        c = matching_circuits[0]
        if i not in circuits[c]:
            circuits[c].append(i)
        if j not in circuits[c]:
            circuits[c].append(j)
        # Merge other circuits into the first one
        for c_idx in reversed(matching_circuits[1:]):
            circuits[c].extend([p for p in circuits[c_idx] if p not in circuits[c]])
            circuits.pop(c_idx)

# Sort circuits by length (longest first)
circuits.sort(key=len, reverse=True)

answer = int(len(circuits[0]))*int(len(circuits[1]))*int(len(circuits[2]))
print('8.1: ' + str(answer))
```

### Day 9
#### Movie Theater
##### part 1
The movie theater has a big tile floor with an interesting pattern.
Some of the tiles are red; the Elves would like to find the largest rectangle that uses red tiles for two of its opposite corners.
A list of where the red tiles are located in the grid (your puzzle input).
Using two red tiles as opposite corners, what is the largest area of any rectangle you can make?

Every red tile is connected to the red tile before and after it by a straight line of green tiles.
All of the tiles inside this loop of red and green tiles are also green.
The rectangle you choose still must have red tiles in opposite corners, but any other tiles it includes must now be red or green. This significantly limits your options.
What is the largest area of any rectangle you can make using only red and green tiles?
